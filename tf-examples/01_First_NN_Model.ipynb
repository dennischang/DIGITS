## Let's build your first Neural Network to predict the numbers.

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

#### Define the Neural Network


```
model = tf.keras.Sequential(
    [keras.layers.Dense(units=1, input_shape=[1])]
    )
```

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

#### Prepare the example data

We follow the formula y = 2x + 1 to prepare the examples data. And let's predict a new number after model trained to see if the model well-trained.

```
x = np.array([1.0, -1.0, 5.0, 2.0, 3.0], dtype=float)
y = np.array([3.0, -1.0, 11.0, 5.0, 7.0], dtype=float)
```

#### Train the Neural Network


```
model.fit(x, y, epochs=500)
```

```
print(model.predict([10.0]))
```
